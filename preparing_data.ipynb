```python
from PIL import Image, ImageDraw
import numpy as np

PATH = r'C:\Users\Lenovo\Desktop\ML\Projekt\Plasmodium'
os.makedirs(PATH, exist_ok=True)
# img_png = glob.glob(PATH + '\*.png') # wszystkie obrazki .png
# img_png_P = glob.glob(PATH + "\RBC" + '\*.png') # wszystkie obrazki P000.png
path_plasmodium = glob.glob(PATH + '\*.jpg')

for f in path_plasmodium:
#     print(f)
    
    # Open the image and ensure RGB
    try:
        img = Image.open(f).convert('RGB')
    except Exception:
        pass
    # Make all background pixels magenta
    ImageDraw.floodfill(img,xy=(0,0),value=(255,0,255),thresh=50)

    # Save for checking
#     im.save('floodfilled.png')

    # Make into Numpy array
    n = np.array(img)

    # Mask of magenta background pixels
    bgMask =(n[:, :, 0:3] == [255,0,255]).all(2)
    count = np.count_nonzero(bgMask)
    size = 10000 #100x100
    if count/size > 0.85:
#         print(f"Background pixels: {count}")
        
        os.remove(f)
```
